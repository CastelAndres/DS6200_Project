```python
import pandas as pd

import sqlite3

nba = sqlite3.connect('nba2024.db')
```

```python
myquery = '''

SELECT *

FROM games

'''
pd.read_sql_query(myquery, con=nba)
```

```python
myquery = '''
SELECT *
FROM games
INNER JOIN teamgames
    ON games.gameid = teamgames.gameid
'''
pd.read_sql_query(myquery, con=nba)
```
